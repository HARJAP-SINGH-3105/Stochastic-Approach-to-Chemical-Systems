```python
import numpy as np
import matplotlib.pyplot as plt
# import mpltern
m=10
mole_fractions = np.zeros((m + 1, 3))
mole_fractions[0] = np.array([0.85, 0.12, 0.03])
volatilities = np.array([2.4, 1, 0.21])
N=1000
n=50

molecules = mole_fractions[0] * N
for step in range(1, m + 1):
    
    for _ in range(n):
        p = molecules * volatilities
        max_p = np.max(p)
        r = p / max_p
        random_numbers = np.random.random()
        molecules = np.where(r > random_numbers, molecules - 1, molecules)
    mole_fractions[step] = molecules / np.sum(molecules)

plt.figure(figsize=(8, 6))
plt.scatter(mole_fractions[:, 1], mole_fractions[:, 2], c=mole_fractions[:, 0], cmap='viridis', s=20)
plt.title('Ternary Diagram')
plt.colorbar(label='Mole Fraction of benzene')
plt.xlabel('Mole Fraction of Toulene')
plt.ylabel('Mole Fraction of cumene')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
for i in range(3):
    plt.plot(range(m + 1), mole_fractions[:, i], label=f'Compound {i+1}')
plt.xlabel('Simulation Step')
plt.ylabel('Mole Fraction')
plt.title('Change of Mole Fraction in Each Step')
plt.legend()
plt.grid(True)
plt.show()
```
