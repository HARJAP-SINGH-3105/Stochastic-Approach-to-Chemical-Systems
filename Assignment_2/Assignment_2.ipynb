```python
import random
import numpy as np3
import matplotlib.pyplot as plt
import plotly.express as px

M = input("Enter initial number of total molecules ")
M = int(M)
x_benzene = input("Enter initial mole fraction of benzene ")
x_toluene = input("Enter initial mole fraction of toluene ")
x_cumene = input("Enter initial mole fraction of cumene ")
x_benzene = float(x_benzene)
x_toluene = float(x_toluene)
x_cumene = float(x_cumene)
M_benzene = x_benzene * M
M_toluene = x_toluene * M
M_cumene = x_cumene * M
a_benzene = input("Enter relative volatility of benzene ")
a_benzene = float(a_benzene)
a_toluene = input("Enter relative volatility of toluene ")
a_toluene = float(a_toluene)
a_cumene = input("Enter relative volatility of cumene ")
a_cumene = float(a_cumene)
m = input("Enter number of steps of distillation ")
m = int(m)
n = input("Enter number of molecules of each type ")
n = int(n)
benzene = []
toluene = []
cumene = []
benzene.append(x_benzene)
toluene.append(x_toluene)
cumene.append(x_cumene)
for i in range(m):
  for j in range(n):
    p_benzene = a_benzene * M_benzene;
    p_toluene = a_toluene * M_toluene;
    p_cumene = a_cumene * M_cumene;
    p_max = max(p_benzene,p_toluene,p_cumene)
    r_benzene = p_benzene / p_max
    r_toluene = p_toluene / p_max
    r_cumene = p_cumene / p_max
    if(float(r_benzene) > random.randrange(0,1)):
      M_benzene -= 1
    if(float(r_toluene) > random.randrange(0,1)):
      M_toluene -= 1
    if(float(r_cumene) > random.randrange(0,1)):
      M_cumene -= 1

    x_benzene = M_benzene / (M_benzene + M_toluene + M_cumene)
    x_toluene = M_toluene / (M_benzene + M_toluene + M_cumene)
    x_cumene = M_benzene / (M_benzene + M_toluene + M_cumene)
    benzene.append(x_benzene)
    toluene.append(x_toluene)
    cumene.append(x_cumene)
    print("Moles of benzene = " + str(M_benzene))
    print("Moles of toluene = " + str(M_toluene))
    print("Moles of cumene = " + str(M_cumene))      

fig=px.scatter_ternary(benzene,toluene,cumene)
#fig=px.scatter_ternary(toluene)
#fig=px.scatter_ternary(cumene)
fig.show()
plt.plot(benzene)
plt.plot(toluene)
plt.plot(cumene)
plt.show()    
```
